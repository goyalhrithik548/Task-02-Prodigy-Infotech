As 'EMBARKED'  was a categorical data we have used 'MODE' to fill null values

## DATA ANALYSIS AND VISUALISATION

```python
# To check the relation between poeple of different'Pclass' who survived

titanic_df.groupby('Pclass')['Survived'].value_counts()
```

```python
# Check the no. of survivals from different 'Pclass'

pclass_survived = titanic_df.groupby('Pclass')['Survived'].sum()
pclass_survived
```

```python
# To get the total number of passengers in different 'Pclass'

pclass_total = titanic_df['Pclass'].value_counts().sort_index()
pclass_total
```

```python
# Create a bar graph for Suvival Rate by 'Pclass'

height = pclass_survived.index 

# plotting the bar using Matplotlib
plt.bar(height,(pclass_survived/pclass_total)*100,color='lightblue')

# Adding labels and title
plt.xlabel('Passenger Class')
plt.ylabel('Survival Percentage')
plt.title('Survival Percentage by Passenger Class')
plt.xticks(height)
```

Our analysis indicates a notable disparity in survival rates among passengers based on their 'Pclass' designation. 

Specifically, individuals in 'Pclass' = 1 exhibited the highest survival rates, while those in 'Pclass' = 3 experienced the lowest survival rates. 

This suggests a direct correlation between 'Pclass' and passenger survival rates.

```python
# Create a new feature 'FamilySize' by combining 'SibSp' and 'Parch'

titanic_df['Family Size'] = titanic_df['SibSp'] + titanic_df['Parch']+1
```

```python
titanic_df.head(15)
```

```python
# To check the relation between different 'Family Sizes' who survived

titanic_df.groupby('Family Size')['Survived'].value_counts()
```

```python
# To create a bar plot using Seaborn to get survival rate wrt Family Size

sns.barplot(x='Family Size',y='Survived',data=titanic_df,hue='Family Size')

# Adding labels and title
plt.xlabel('Family Size')
plt.ylabel('Survival Percentage')
plt.title('Survival Percentage by Family Members')
```

Here, our analysis reveals notable trends regarding the relationship between family size and survival likelihood. Specifically, passengers with a family size of 4 exhibited the highest probability of survival, whereas those with family sizes of 8 and 11 experienced the lowest survival rates. Notably, individuals traveling with three family members had the highest chances of survival.

```python
# To create survival rate by different age group

titanic_df['AgeGroup'] = pd.cut(titanic_df['Age'], bins=[0, 18, 30, 50,90], labels=['Child', 'Young Adult', 'Adult', 'Senior'])
sns.barplot(x='AgeGroup', y='Survived', data=titanic_df)
plt.title('Survival Rate by Age Group')
```

Created a new column 'AgeGroup' which categorize the age data in the following aspects:-

We have enriched our dataset by creating a new column named 'AgeGroup', which provides a categorical representation of the age data. This segmentation offers a more nuanced understanding of age demographics, aiding in the analysis and interpretation of various patterns and trends within the dataset. By categorizing ages into distinct groupsChild, Young Adult, Adult, and Oldwe have effectively organized the data, allowing for clearer insights into age-related phenomena. This approach facilitates more targeted and insightful exploratory data analysis, enabling researchers and analysts to uncover valuable insights with greater precision and depth.
